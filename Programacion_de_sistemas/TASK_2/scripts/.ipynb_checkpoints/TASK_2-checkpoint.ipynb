**Resolve the matrix equation with numpy functions**

    1.Loading A and b from a file on disk.
        The file can be generated by running the Python program
        equation-system-generator.py
    2.Solving the equation, i.e. obtaining the values for x i
    3.Show the results in order to check if they match with the solution
    provided.

```python
#Libraries
import numpy
import MyEquation


#Load data and resolve de the equations
def MyEquations():
    A,b = MyEquation.load_equation_system( 'system.txt' )
    #This line solve the equation system
    x=numpy.linalg.solve(A,b)
    test=numpy.matmul(A,x) 
    return A,b,x,test

#Test prints to evaluate the results
def Myprints(A,b,x,test):
    
    # Here you have to solve A * x = b, look at in the linear algebra package of numpy for the solution
    print('The equation is A*x=b')
    print('\nThis is A: \n\n' + str(A))
    print('\nThis is b: \n\n' + str(b))
    print ('\nThis is x using the numpy function \'numpy.linalg.solve\': \n\n' + str(x))
    print ('\nTo test if this result is correct now we multiply A*x and the result should be b')
    print ('\nThis is the test with the numpy function \'numpy.matmul\': \n\n' + str(test))
    print ('\nNow with \'numpy.isclose\' we can compare test and b: \n\n' + str(numpy.isclose(test,b)))

#Main function    
def main():
    A,b,x,test=MyEquations()
    Myprints(A,b,x,test)

#GO!    
if __name__ == "__main__":
    main()
    
```
